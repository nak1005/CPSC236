### Specifications:
- Use a list of lists to store the data for the contacts. Provide starting data for two or more contacts.
- For the `view` and `del` commands, display an error message if the user enters and invalid contact number.
- When you exit the program, all changes that you made to the contact list are lost.




````python
def menuDisplay():
    print("COMMAND MENU")
    print("list - Display all contacts\n" +
            "view - View a contact\n" +
            "add - Add a contact\n" +
            "del - Delete a contact\n" +
            "exit - Exit program")
def getCommand(contacts):
    while True:
        command = input("\nCommand: ")
        if command == 'list':
            listDisplay(contacts)
        elif command == 'view':
            viewContact(contacts)
        elif command == 'add':
            addContact(contacts)
        elif command == 'del':
            delContact(contacts)
        elif command =='exit':
            break
        else:
            print("Invalid command. Try again!")        
def listDisplay(contacts):
    for i in range(0, len(contacts)):
        print(i+1, ".", contacts[i][0]) 
def viewContact(contacts):
    contactNum = int(input("Number: "))
    if contactNum < 1 or contactNum > len(contacts):
        print("That contact doesn't exist!\n")
    else:
        for i in range(0, len(contacts)):
            if contactNum == i+1:
                print("Name:", contacts[i][0], 
                      "\nEmail:", contacts[i][1], 
                      "\nPhone:", contacts[i][2]) 
def addContact(contacts):
    name = input("Name: ")
    email = input("Email: ")
    phone = input("Phone: ")
    contact = []
    contact.append(name)
    contact.append(email)
    contact.append(phone)
    contacts.append(contact)
    print(name + " was added.")
def delContact(contacts):
    contactNum = int(input("Number: "))
    if contactNum < 1 or contactNum > len(contacts):
        print("That contact doesn't exist!\n")
    else:
        print(contacts[contactNum-1][0], "was deleted.")
        contacts = contacts.pop(contactNum-1)
def exitMenu():
    choice = 'n'
    return choice
        
def main():
    contacts = [["Guido van Rossum", "guido@guidovanrossum.com", "559-123-4568"], 
                ["Eric Idle", "eric@ericidle.com", "+44 20 7946 0958"]]
    print("Contact Manager\n")
    menuDisplay()
    getCommand(contacts)
    print("Bye!")
    
#calls main()
if __name__ == "__main__":
    main()
````
